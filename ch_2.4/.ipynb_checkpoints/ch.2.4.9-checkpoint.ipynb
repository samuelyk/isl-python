```python
# 9.(e)
# Using the full data set, investigate the predictors graphically,
# using scatterplots or other tools of your choice. Create some plots
# highlighting the relationships among the predictors. 
# Comment on your findings.
```

```python
# Reset Index

Auto = Auto.set_index('name')
Auto.head()
```

```python
# Convert to categorical variables
# cylinders, year, origin

for col in ['cylinders', 'year', 'origin']:
    print(f"\'{col}\' orignial data type: {Auto[col].dtype}")

    Auto[col] = Auto[col].astype('category')
    print(f"\'{col}\' new data type: {Auto[col].dtype}\n")
```

```python
# Boxplot of MPG, Horsepower, and Acceleration against the categorical variables

fig, axes = plt.subplots(1, 3, figsize=(15,5))
Auto.boxplot('mpg', by='cylinders', ax=axes[0])
Auto.boxplot('mpg', by='year', ax=axes[1])
Auto.boxplot('mpg', by='origin', ax=axes[2])
fig.suptitle('Boxplot of MPG vs Categorical Variables');

fig, axes = plt.subplots(1, 3, figsize=(15,5))
Auto.boxplot('horsepower', by='cylinders', ax=axes[0])
Auto.boxplot('horsepower', by='year', ax=axes[1])
Auto.boxplot('horsepower', by='origin', ax=axes[2])
fig.suptitle('Boxplot of Horsepower vs Categorical Variables');

fig, axes = plt.subplots(1, 3, figsize=(15,5))
Auto.boxplot('acceleration', by='cylinders', ax=axes[0])
Auto.boxplot('acceleration', by='year', ax=axes[1])
Auto.boxplot('acceleration', by='origin', ax=axes[2])
fig.suptitle('Boxplot of Acceleration vs Categorical Variables');
```

```python
# Histogram of mpg, displacement, horsepower, weight, acceleration
# Insert mean of each variable with red line

fig, axes = plt.subplots(2, 3, figsize=(15,8))

Auto.hist('mpg', ax=axes[0,0])
axes[0,0].axvline(Auto['mpg'].mean(), color='red', linestyle='--', linewidth=2)

Auto.hist('displacement', ax=axes[0,1])
axes[0,1].axvline(Auto['displacement'].mean(), color='red', linestyle='--', linewidth=2)

Auto.hist('horsepower', ax=axes[0,2])
axes[0,2].axvline(Auto['horsepower'].mean(), color='red', linestyle='--', linewidth=2)

Auto.hist('weight', ax=axes[1,0])
axes[1,0].axvline(Auto['weight'].mean(), color='red', linestyle='--', linewidth=2)

Auto.hist('acceleration', ax=axes[1,1])
axes[1,1].axvline(Auto['acceleration'].mean(), color='red', linestyle='--', linewidth=2)

axes[1, 2].axis('off'); # Hide the last blank plot

# Get mean values for the quantitative variables

quantitative_mean = Auto[['mpg', 'displacement', 'horsepower', 'weight', 'acceleration']]\
                    .mean().to_frame(name='Mean').T
quantitative_mean.columns.name = 'Quantitative Variables'
quantitative_mean
```

```python
# Scatterplot Matrix for the quantitative variables

pd.plotting.scatter_matrix(Auto[['mpg', 'displacement', 'horsepower', 'weight', 'acceleration']], figsize=(8,8));
```
