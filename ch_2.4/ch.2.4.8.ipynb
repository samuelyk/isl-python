```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
college = pd.read_csv('College.csv')
```

```python
college
```

```python
college.info()
```

```python
# Set the 1st col as an index
# Method 1

college_a = pd.read_csv('College.csv', index_col=0)
college_a.head()
```

```python
# Set the 1st col as an index
# Method 2

college_b = college.rename({'Unnamed: 0': 'College'}, axis=1)
college_b = college_b.set_index('College')
college_b.head()
```

```python
# We will use college_b going forward

college = college_b
```

```python
# Numerical summary of college

college.describe()
```

```python
pd.plotting.scatter_matrix(college[['Top10perc', 'Apps', 'Enroll']]);
```

```python
# Convert 'Private' variable to categorical

print(college['Private'].dtype)

college['Private'] = college['Private'].astype('category')
college['Private'].dtype
```

```python
# Boxplot of 'Outstate' vs 'Private'

fig, ax = plt.subplots(figsize=(8,8))
college.boxplot('Outstate', by='Private', ax=ax);
```

```python
# Create a new categorical variable 'Elite'
# 'Elite': Yes if the proportion of students coming from the top 10%
# of their high school classes exceeds 50%

college['Prop of Top10perc'] = college['Top10perc'] / college['Enroll']
print(college['Prop of Top10perc'])

college['Elite'] = pd.cut(college['Prop of Top10perc'], [0, 0.5, 1], labels=['No', 'Yes'])
```

```python
college['Elite']
```

```python
college['Elite'].value_counts()
```

```python
college[college['Elite'] == 'Yes'][['Enroll', 'Top10perc', 'Prop of Top10perc']]
```

```python
# Boxplot of 'Outstate' vs 'Elite'

fig, ax = plt.subplots(figsize=(8,8))
college.boxplot('Outstate', by='Elite', ax=ax);
```

```python
# Histograms

fig, axes = plt.subplots(2, 2, figsize=(12,5))
college.hist('Room.Board', ax=axes[0,0])
college.hist('Outstate', ax=axes[0,1])
college.hist('Top10perc', ax=axes[1,0])
college.hist('Top25perc', ax=axes[1,1])
```

```python
# Scatter plots - 1
# Does the number of top students affect 'Grad.Rate'?
# Having more Ph.D faculties lead to higher 'Grad.Rate'?
# Does alumni donation help 'Grad.Rate'?

fig, axes = plt.subplots(2, 2, figsize=(16,16))
college.plot.scatter('Top10perc', 'Grad.Rate', ax=axes[0,0]);
college.plot.scatter('Top25perc', 'Grad.Rate', ax=axes[0,1]);
college.plot.scatter('PhD', 'Grad.Rate', ax=axes[1,0]);
college.plot.scatter('perc.alumni', 'Grad.Rate', ax=axes[1,1]);
```

```python
# Scatter plots - 2
# If colleges have higher Out-of-state tuition, will they chrage less on room/board costs?
# If colleges have higher Out-of-state tuition, will there be lower student/faculty ratio?
# If colleges have higher Out-of-state tuition, do they spend more on instructional expenditure?

fig, axes = plt.subplots(1, 3, figsize=(20,5))
college.plot.scatter('Outstate', 'Room.Board', ax=axes[0]);
college.plot.scatter('Outstate', 'S.F.Ratio', ax=axes[1]);
college.plot.scatter('Outstate', 'Expend', ax=axes[2]);
```

```python
# Boxplots

fig, axes = plt.subplots(3, 3, figsize=(30,30))
college.boxplot('Apps', by='Private', ax=axes[0,0])
college.boxplot('Room.Board', by='Private', ax=axes[0,1])
college.boxplot('Books', by='Private', ax=axes[0,2])
college.boxplot('Personal', by='Private', ax=axes[1,0])
college.boxplot('PhD', by='Private', ax=axes[1,1])
college.boxplot('S.F.Ratio', by='Private', ax=axes[1,2])
college.boxplot('perc.alumni', by='Private', ax=axes[2,0])
college.boxplot('Expend', by='Private', ax=axes[2,1])
college.boxplot('Grad.Rate', by='Private', ax=axes[2,2]);
```
