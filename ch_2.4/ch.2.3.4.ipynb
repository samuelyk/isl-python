```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
rng = np.random.default_rng(234)
x = rng.standard_normal(100)
y = rng.standard_normal(100)
```

```python
# subplots() returns figure object and axes object(s)
# ax.plot() defaults to plot a line plot

fig, ax = plt.subplots(figsize=(8, 8))
ax.plot(x, y)
```

```python
# Plot a scatterplot by passing 'o' argument to ax.plot()

fig, ax = plt.subplots(figsize=(8, 8))
ax.plot(x, y, 'o')
```

```python
# Alternative method for scatterplot with ax.scatter()
# semicolon at the end the last line prevents printing out text.

fig, ax = plt.subplots(figsize=(8,8))
ax.scatter(x, y, marker='+');
```

```python
# Labeling plot

fig, ax = plt.subplots(figsize=(8,8))
ax.plot(x, y, '*')
ax.set_xlabel("This is the x-axis")
ax.set_ylabel("This is the y-axis")
ax.set_title("Plot of X vs Y");
```

```python
# Access fig object and change the size

fig.set_size_inches(12, 3)
fig
```

```python
# Create six plots within a figure
# Passing sharex=True kwarg will make all plots have a common x-axis 

fig, axes = plt.subplots(nrows=2, ncols=3, figsize=(15,5))
```

```python
axes[0,1].plot(x, y, 'o')
axes[1,2].scatter(x, y, marker='x')
fig
```

```python
# Saving figure
# dpi: dots per inch

fig.savefig("Six_plots_in_one_figure.png", dpi=400)
fig.savefig("Six_plots_in_one_figure.pdf", dpi=200)
```
