```python
import matplotlib.pyplot as plt
import pandas as pd

Auto = pd.read_csv('Auto.csv', na_values=['?']) # Replace '?' to NaN
Auto = Auto.dropna()
```

```python
# Create a scatter plot 'horsepower' vs 'mpg'

fig, ax = plt.subplots(figsize=(8,8))
ax.scatter(Auto['horsepower'], Auto['mpg'], marker= '+');
```

```python
# Alternative method: Auto.plot()

ax = Auto.plot.scatter('horsepower', 'mpg')
ax.set_title('Horsepower vs MPG');
```

```python
# Save the figure with ax

# ax.figure.savefig('horsepower_mpg.png')
```

```python
fig, axes = plt.subplots(ncols=3, figsize=(12,5))
Auto.plot.scatter('horsepower', 'mpg', ax=axes[1]);
```

```python
# Replace quantitative 'cylinders' variable to categorical (qualitative) variable

print(sorted(Auto.cylinders.unique())) # Note columns of a df can be accessed as attributes

Auto.cylinders = pd.Series(Auto.cylinders, dtype='category')
Auto.cylinders.dtype
```

```python
# Now 'cylinders' is qualitative
# Display boxplot for 'cylinders'

fig, ax = plt.subplots(figsize=(8,8))
Auto.boxplot('mpg', by='cylinders', ax=ax);
```

```python
# Plot a histogram

fig, ax = plt.subplots(figsize=(8,8))
Auto.hist('mpg', ax=ax);
```

```python
# Change bar color and the number of bins

fig, ax = plt.subplots(figsize=(8,8))
Auto.hist('mpg', color='red', bins=12, ax=ax); # Default bin numbers: 10
```

```python
# Create a scatterplot matrix

pd.plotting.scatter_matrix(Auto);
```

```python
# scatterplots for a subset of the variables

pd.plotting.scatter_matrix(Auto[['mpg', 'displacement', 'weight']]);
```

```python
# Show a numerical summary of each column

Auto.describe()
```

```python
# Numerical summary for one column

Auto['mpg'].describe()
```

```python
# Numberical summary for multiple columns

Auto[['mpg', 'weight']].describe()
```
