```python
# Replace '?' to np.nan
# Notice dtype of Auto['horsepower']

Auto = pd.read_csv('Auto.data',
                   na_values=['?'],
                   sep = r'\s+')
Auto['horsepower']
```

```python
np.unique(Auto['horsepower'])
```

```python
# Now we can perform numerical operations

print(Auto['horsepower'].sum())
```

```python
# How many rows and columns does Auto have?

print(Auto.shape)
print(f"It has {Auto.shape[0]} rows and {Auto.shape[1]} columns.")
```

```python
# Count missing observaions per columns
# Method 1 - df.isna().sum()

Auto.isna().sum()
```

```python
# Method 2 - df.isnull().sum()

Auto.isnull().sum()
```

```python
# Remove rows with those missing observations

Auto_new = Auto.dropna()
print(Auto_new.shape)
print(Auto_new.isna().sum())
```

```python
# Basics of Selecting Rows and Columns

Auto = Auto_new # overwrite the previous value
Auto.columns
```

```python
# Get the first three rows

Auto[:3]
```

```python
# Boolean indexing

idx_80 = Auto['year'] > 80
Auto[idx_80]
```

```python
Auto[['mpg', 'horsepower']]
```

```python
# Current index of Auto data frame

Auto.index
```

```python
# Rename index to the contents of 'name' column

Auto_re = Auto.set_index('name')
Auto_re
```

```python
Auto_re.index
```

```python
# 'name' column is no longer in the Auto columns

Auto_re.columns
```

```python
# loc[]
# Access rows with the name of the index

Auto_re.loc[['amc rebel sst', 'ford torino']]
```

```python
# iloc[]
# Access rows with the integer location of the index (0-based)

Auto_re.iloc[[3,4]]
```

```python
Auto_re.iloc[:,[0,2,3]]
```

```python
Auto_re.iloc[[3,4],[0,2,3]]
```

```python
# Duplicate index: ford galaxie 500

Auto_re.loc['ford galaxie 500', ['mpg', 'origin']]
```

```python
# More on Selecting Rows and Columns
# A subset with 'year' > 80 & columns 'weight' & 'origin'

idx_80 = Auto_re['year'] > 80
Auto_re.loc[idx_80,['weight', 'origin']]
```

```python
# Another method to get the same subset

Auto_re[Auto_re['year'] > 80][['weight', 'origin']]
```

```python
# Another method with lambda function
# lambda call creates a function that takes a single argument (df),
# and returns df['year'] > 80.

Auto_re.loc[lambda df: df['year'] > 80, ['weight', 'origin']]
```

```python
# More lambda example - 1

Auto_re.loc[lambda df: (df['year'] > 80) & (df['mpg'] > 30),
            ['weight', 'origin']]
```

```python
# More lambda example - 2

Auto_re.loc[lambda df: (df['displacement'] < 300) 
                        & (df.index.str.contains('ford') 
                        | (df.index.str.contains('datsun')),
                        ['weight', 'origin']]
```

```python
# Note:
# For integer based queries, use the iloc[] method.
# For string and Boolean selections, use the loc[] method.
# For functional queries that filter rows, use the loc[] method 
# with a function (typically a lambda) in the rows argument.
```
